### Step 4. Create a histogram with the 10 countries that have the most 'Quantity' ordered except UK

```python
no_UK = online_rt[online_rt.Country != 'United Kingdom']
no_UK
```

```python
top_10 = no_UK.groupby('Country').agg({'Quantity':'sum'}).sort_values('Quantity',ascending = False)[ : 10]
top_10
```

```python
top_10.plot(kind = 'bar',figsize = (8,8))
```
